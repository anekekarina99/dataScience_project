```python
import random
import pandas as pd

# Dataset awal (gunakan data dari contoh sebelumnya)
data = {
    'CUSTID': [101, 102, 103, 104, 105],
    'BALANCE': [1000, 2000, 1500, 3000, 2500],
    'BALANCEFREQUENCY': [0.8, 1.0, 0.6, 0.9, 0.7],
    'PURCHASES': [500, 1000, 800, 1200, 900],
    'ONEOFFPURCHASES': [200, 600, 400, 800, 500],
    'INSTALLMENTSPURCHASES': [300, 400, 400, 400, 400],
    'CASHADVANCE': [100, 0, 500, 0, 200],
    'PURCHASESFREQUENCY': [0.6, 1.0, 0.4, 0.8, 0.5],
    'ONEOFFPURCHASESFREQUENCY': [0.4, 0.8, 0.6, 1.0, 0.5],
    'PURCHASESINSTALLMENTSFREQUENCY': [0.5, 0.8, 0.8, 0.8, 0.8],
    'CASHADVANCEFREQUENCY': [0.2, 0.0, 0.4, 0.0, 0.3],
    'CASHADVANCETRX': [2, 0, 3, 0, 1],
    'PURCHASESTRX': [10, 15, 12, 20, 14],
    'CREDITLIMIT': [5000, 8000, 7000, 10000, 9000],
    'PAYMENTS': [800, 1500, 1000, 1800, 1200],
    'MINIMUM_PAYMENTS': [300, 500, 400, 600, 450],
    'PRCFULLPAYMENT': [0.3, 0.5, 0.2, 0.6, 0.4],
    'TENURE': [5, 6, 4, 7, 6]
}

# Membuat DataFrame dari data
df = pd.DataFrame(data)

# Fungsi untuk membuat dataset baru secara acak berdasarkan data yang ada
def augment_data(data, n_samples):
    augmented_data = []
    for _ in range(n_samples):
        random_data = {}
        for col in df.columns:
            random_data[col] = random.choice(df[col])
        augmented_data.append(random_data)
    return augmented_data

# Augmentasi 100 data baru
augmented_data = augment_data(df, 100)

# Membuat DataFrame baru dari data augmentasi
augmented_df = pd.DataFrame(augmented_data)

# Menampilkan dataset hasil augmentasi
print(augmented_df)
```
